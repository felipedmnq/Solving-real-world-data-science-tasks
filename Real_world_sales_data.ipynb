# Question 4: What products are most often sold together?

```python
df_sales.head()
```

```python
df_orders = df_sales[df_sales['Order_ID'].duplicated(keep = False)]
```

```python
df_orders.head()
```

```python
# Grouping products by order ID - Keith's function.

df_orders['grouped'] = df_orders.groupby('Order_ID')['Product'].transform(lambda x: ','.join(x))
```

```python
df_orders.head()

# The columns Product and grouped show us the same, so we need to drop some rows.
```

```python
df_orders = df_orders[['Order_ID', 'grouped']].drop_duplicates()
```

```python
df_orders
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()

for row in df_orders['grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
count.most_common(10)
```

```python
for key, value in count.most_common(10):
    print(key, value)
```
